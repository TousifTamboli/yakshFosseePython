```python
import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import interact

def damp(alpha):
    t = np.linspace(0, 4 * np.pi, 200)  # Time array from 0 to 4*pi, with 200 points
    plt.plot(t, np.exp(-t * alpha) * np.sin(t))  # Damped sine wave equation
    plt.xlabel('Time')
    plt.ylabel('Amplitude')
    plt.title(f'Damped Sine Wave with alpha = {alpha}')
    plt.grid(True)
    plt.show()

# Create an interactive widget for the alpha parameter
interact(damp, alpha=(0, 2, 0.1))  # alpha goes from 0 to 2 with step size 0.1
```

```python
def damp(alpha, beta):
    t = linspace(0, 4*pi, 200)
    plot(t, exp(-t*alpha)*sin(t+beta))

interact(damp, alpha=0.2, beta=5.0);
```
